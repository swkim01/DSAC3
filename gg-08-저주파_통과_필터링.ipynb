```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 이미지 읽기
!curl -L https://github.com/swkim01/DSAC5/raw/main/data/case.jpg \
    -o case.jpg
img = cv2.imread('./case.jpg')
```

```python
kernel = np.ones((5,5), np.float32) / 25
dst = cv2.filter2D(img, -1, kernel)
```

```python
plt.figure(figsize=(15,15))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('blur')
plt.xticks([]), plt.yticks([])
plt.show()
```

<h1>가우시안 필터링(Gaussian Filtering)</h1>

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('./case.jpg')
```

```python
blur = cv2.GaussianBlur(img,(5,5),0)
```

```python
plt.figure(figsize=(15,15))
plt.subplot(121),plt.imshow(img), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]),plt.yticks([])
plt.show()
```

<h1> Median Filtering</h1>


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('./case.jpg')
```

```python
blur = cv2.medianBlur(img,5)
```

```python
plt.figure(figsize=(15,15))
plt.subplot(121), plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show
```

<h1>Bilateral Filtering</h1>

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('./case.jpg')
```

```python
blur = cv2.bilateralFilter(img,9,75,75)
```

```python
plt.figure(figsize=(15,15))
plt.subplot(121), plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show
```
